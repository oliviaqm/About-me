## Cookie cutter code for extracting comments from a YouTube video

```python
import googleapiclient.discovery
import googleapiclient.errors

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "AIzaSyCXMv_hyOmrrDziGJX31zVHhCObu0WJpc4"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey = DEVELOPER_KEY
)

# Test: extract comments on Hey! Say! JUMP - 震災復興支援パフォーマンス - ヒカリサス video
request = youtube.commentThreads().list(
    part = "snippet",
    videoId = "kK_ByPA08Hw", # videoID is the part of the video's URL, after "https://www.youtube.com/watch?". e.g."kK_ByPA08Hw" in https://www.youtube.com/watch?kK_ByPA08Hw
    maxResults=100
)

response = request.execute()

i = 0
for item in response['items']:
  print(i)
  print(item['snippet']['topLevelComment']['snippet']['textDisplay'])
  i += 1
```
